```python
import sqlite3 as dblite
```

```python
conn=dblite.connect("mci.db")
cur = conn.cursor()
```

```python
mci_data = open("mci.txt","r")

for doc_data in mci_data.readlines():
    doc_data=doc_data[:-1]
    doc_data=doc_data.split(",")
    if len(doc_data) !=6:
        print(doc_data)
    print(doc_data)
    cur.execute("insert into MCi values ((?),(?),(?),(?),(?),(?))",doc_data)
    conn.commit()
cur.close()
conn.close()
        
```
